```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import re #usamos la libreria re de regular expression
```

```python
webpage= requests.get("https://footballdatabase.com/league-scores-tables/argentina-primera-division-2022")
soup = BeautifulSoup(webpage.content, "html.parser") #todo el codigo html de la pagina
```

```python
soup.find(id='total')
liga= soup.find_all(id='total') # identificamos la tabla y buscamos y extraemos ella del codigo html
```

```python
liga= str(liga) # transformamos a str la tabla
```

```python
liga= re.sub(r'<.*?>', lambda g: g.group(0).upper(), liga) #cambiamos a mayusculas
liga
```

```python
df = pd.read_html(liga)[0] #usamos con el texto en string un read html de pandas
df # nos genera el dataframe
```

```python
df.describe().T
```

```python
df.describe(include='O').T
```

```python
# cambiamos los nombres de las columnas para que se entienda mejor la tabla
df= df.rename(columns={'#':'Posición', 'P': 'PUNTOS', '+/-': 'DIF GOL','M':'PARTIDOS JUGADOS','W':'GANADOS',
                       'D':'EMPATES','L':'PERDIDOS','F':'GOLES A FAVOR','A':'GOLES ENCONTRA'})
df            
```

```python
df= df.set_index('Posición')
```
