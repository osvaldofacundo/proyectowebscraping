Creamos dos columnas con goles de local y vistante, a partir de la columna resultado.

```python
#creamos dos columnas una para gol visitante y otra para gol local seprando desde la columna Resultado tomando patron -
#tomando solo dos divisiones y con la opción expand true
fixture2022[['gol_local', 'gol_visitante']]= fixture2022.Resultado.str.split('-',2, expand=True)
fixture2022
```

nos quedaron salvo fecha del campeonato, todas las columnas como object - tanto gol de visitante y local deben de ser int64. Nos encontramos con el problema de que existen dos partidos aplazados y por ello no podemos cambiar el valor a númérico

```python
aplazado= fixture2022[fixture2022.loc[:,'gol_local']== 'Aplazado'].index
fixture2022= fixture2022.drop(aplazado)
```

```python
fixture2022
```

```python
fixture2022['gol_local'] = fixture2022['gol_local'].astype(np.int64) #le asignamos tipo valor float
fixture2022['gol_visitante'] = fixture2022['gol_visitante'].astype(np.int64)
fixture2022
```

```python
fixture2022.dtypes
```

Debemos ahora crear una columna que nos indique el resultado. esto es, si gana el local, el visitante o si empataron

```python
#definimos las condiciones que deben darse para uno u otro resultado

condiciones= [fixture2022['gol_local']>fixture2022['gol_visitante'], fixture2022['gol_local']<fixture2022['gol_visitante']]

# definimos las opciones
opciones= ['local', 'visitante']

#creamos la columna del dataframe que muestre el resultado de la comparación
fixture2022['resultado'] =np.select(condiciones, opciones, 'empate')
```

```python
fixture2022= fixture2022.rename(columns={'Resultado':'marcador'}) #cambiamos la columna del marcador para no confundir el resultado
fixture2022
```
